```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Cargar datos
df_encuesta = pd.read_excel('Encuesta.xlsx')
df_plataforma = pd.read_excel('Datos plataforma.xlsx')

# Selección de variables necesarias y eliminación de valores nulos
datos = pd.DataFrame({'Peso': df_encuesta['Peso (kg)'],'Presion_Pie_Izq': df_plataforma['Presión promedio en el pie Izq. (gr/cm²)'],'Presion_Pie_Der': df_plataforma['Presión promedio en el pie Der. (gr/cm²9']}).dropna()

# Función para realizar el análisis y mostrar interpretación automática
def analizar_correlacion_peso_presion(peso, presion, pie):
    # Cálculo de la correlación de Pearson
    corr, p_valor = pearsonr(peso, presion)
    print(f"\nCorrelación entre Peso y Presión promedio en el Pie {pie}:")
    print(f"Coeficiente de correlación: {corr:.4f}")
    print(f"Valor p: {p_valor:.4f}")

    # Interpretación
    if p_valor < 0.05:
        interpretacion = (
            f"Existe una correlación significativa entre el peso y la presión promedio en el pie {pie} "
            f"con un coeficiente de correlación de {corr:.4f}. "
            f"{'Esta es una correlación positiva' if corr > 0 else 'Esta es una correlación negativa'}, lo cual indica que "
            f"{'a mayor peso, la presión tiende a aumentar' if corr > 0 else 'a mayor peso, la presión tiende a disminuir'} en el pie {pie}."
        )
    else:
        interpretacion = (
            f"No se encontró una correlación significativa entre el peso y\nla presión promedio en el pie {pie} "
            f"(coeficiente de correlación: {corr:.4f}, valor p: {p_valor:.4f}).\n"
            "Esto sugiere que no existe una relación lineal clara entre el peso y la presión en este pie."
        )
    print(interpretacion)

    # Gráfico de dispersión con línea de regresión
    plt.figure(figsize=(10, 5))
    sns.regplot(x=peso, y=presion, scatter_kws={'s': 50}, line_kws={'color': 'red'})
    plt.xlabel('Peso (kg)')
    plt.ylabel(f'Presión promedio en el Pie {pie} (gr/cm²)')
    plt.title(f'Relación entre Peso y Presión promedio en el Pie {pie}')
    plt.show()

# Realizar análisis para cada pie
analizar_correlacion_peso_presion(datos['Peso'], datos['Presion_Pie_Izq'], 'Izquierdo')
analizar_correlacion_peso_presion(datos['Peso'], datos['Presion_Pie_Der'], 'Derecho')


```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Cargar datos
df_encuesta = pd.read_excel('Encuesta.xlsx')
df_plataforma = pd.read_excel('Datos plataforma.xlsx')

# Cargar datos de dificultades para caminar y carga en el antepié y retropié
datos_dificultad_carga = pd.DataFrame({'Dificultades_Caminar': df_encuesta['¿Tiene dificultades para caminar? '],'Carga_Antepie_Izq': df_plataforma['Carga en el antepié Izq. (%)'],'Carga_Antepie_Der': df_plataforma['Carga en el antepié Der. (%)'],'Carga_Retropie_Izq': df_plataforma['Carga en el retropié Izq. (%)'],'Carga_Retropie_Der': df_plataforma['Carga en el retropié Der. (%)']}).dropna()

# Limpiar espacios adicionales en las respuestas de dificultades para caminar
datos_dificultad_carga['Dificultades_Caminar'] = datos_dificultad_carga['Dificultades_Caminar'].str.strip()

# Modificar las etiquetas de "Dificultades para Caminar" para incluir saltos de línea
datos_dificultad_carga['Dificultades_Caminar'] = datos_dificultad_carga['Dificultades_Caminar'].replace({
    'Sí, pero solo en superficies irregulares': 'Sí, en superficies\nirregulares',
    'Sí, también en distancias cortas': 'Sí, en distancias\ncortas',
    'No tengo ninguna dificultad': 'No tengo\nninguna dificultad',
    'Sí, al caminar distancias largas': 'Sí, en distancias\nlargas'
})
# Definir una función para realizar el ANOVA y graficar
def analizar_y_graficar_carga(datos, dificultad_col, carga_col, carga_desc):
    niveles_dificultad = datos[dificultad_col].unique()
    grupos_carga = [datos[datos[dificultad_col] == nivel][carga_col] for nivel in niveles_dificultad]

    # Realizar ANOVA
    f_stat, p_valor = f_oneway(*grupos_carga)
    print(f"Análisis de Dificultades para Caminar vs. {carga_desc}")
    print(f"Estadístico F: {f_stat:.4f}, Valor p: {p_valor:.4f}")

    # Interpretación del resultado de ANOVA
    if p_valor < 0.05:
        interpretacion = f"Existe una diferencia significativa en la {carga_desc.lower()} según las dificultades para caminar."
    else:
        interpretacion = f"No se encontró una diferencia significativa en la {carga_desc.lower()} según las dificultades para caminar."
    print(interpretacion)

    # Gráfico de caja y bigotes
    plt.figure(figsize=(12, 6))
    ax = sns.boxplot(x=dificultad_col, y=carga_col, data=datos)
    plt.xlabel('Dificultades para Caminar')
    plt.ylabel(f'{carga_desc} (%)')
    plt.title(f'{carga_desc} según Dificultades para Caminar')

    # Añadir etiquetas de la mediana en cada caja
    for i, nivel in enumerate(niveles_dificultad):
        mediana = datos[datos[dificultad_col] == nivel][carga_col].median()
        ax.annotate(f'{mediana:.2f}', xy=(i, mediana), ha='center', va='center', fontsize=10, color='black', fontweight='bold', xytext=(0, 5), textcoords='offset points')
    plt.show()

# Ejecutar el análisis y gráfico para cada una de las cargas
analizar_y_graficar_carga(datos_dificultad_carga, 'Dificultades_Caminar', 'Carga_Antepie_Izq', 'Carga en el Antepié Izquierdo')
analizar_y_graficar_carga(datos_dificultad_carga, 'Dificultades_Caminar', 'Carga_Antepie_Der', 'Carga en el Antepié Derecho')
analizar_y_graficar_carga(datos_dificultad_carga, 'Dificultades_Caminar', 'Carga_Retropie_Izq', 'Carga en el Retropié Izquierdo')
analizar_y_graficar_carga(datos_dificultad_carga, 'Dificultades_Caminar', 'Carga_Retropie_Der', 'Carga en el Retropié Derecho')
```
